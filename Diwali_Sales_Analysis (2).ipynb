# Exploratory Data Analysis

# Gender

```python
df.columns
```

```python
#countplot is used to count the data and x=Gender will put the value in x-axis and there corresponding values on Y-axis
#data=df here the data is retrieved from the name of dataframe i.e., df
ax = sns.countplot(x='Gender', data=df)

for bars in ax.containers:
    ax.bar_label(bars)
#this is used to add labels to the charts, it will also gives the value of each data on top of the containers
```

```python
df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
```

```python
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x='Gender', y='Amount', data = sales_gen)
```

From above graph we can observe that most of the buyers are females and even the purchasing power of females are greater than male

# Age

```python
df.columns
```

```python
sns.countplot( data=df, x='Age Group', hue = 'Gender')
#gives more detail graph differentiating the type of buyer with age group
```

```python
df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
```

```python
ax = sns.countplot( data=df, x='Age Group', hue = 'Gender')

for bars in  ax.containers:
    ax.bar_label(bars)
```

From above graph we can observe that most of the buyers are of age group between 26-35 years that to females

# State

```python
df.columns
```

```python
df.State
```

```python
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10) 
#.head() shows the top values of the graph

sns.set(rc={'figure.figsize':(16, 5)}) #it gives the size of the graph
sns.barplot (data=sales_state, x = 'State', y='Orders')
```

From above graphs we can see that most of the orders are from Uttar Pradesh, Maharashtra and Karnataka 

# Marital Status

```python
ax = sns.countplot(x='Marital_Status', data = df)
sns.set(rc={'figure.figsize':(7, 5)}) #it gives the size of the graph
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(6, 5)}) #it gives the size of the graph
sns.barplot (data=sales_state, x = 'Marital_Status', y='Amount', hue = 'Gender')
```

We can see that most of the buyers are married women

# Occupation

```python
ax = sns.countplot(x='Occupation', data = df)
sns.set(rc={'figure.figsize':(20, 5)}) #it gives the size of the graph
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(20, 5)}) #it gives the size of the graph
sns.barplot (data=sales_state, x = 'Occupation', y='Amount')
```

IT sector have high purchasing power followed by Healthcare and Aviation Sector 

# Product cateogary

```python
sns.set(rc={'figure.figsize':(22, 5)}) #it gives the size of the graph
ax = sns.countplot( data = df, x='Product_Category')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20, 5)}) #it gives the size of the graph
sns.barplot (data=sales_state, x = 'Product_Category', y='Amount')
```

From above graph we can see that most amount is spent on FOOD, CLOYHING and ELECTRONIC
